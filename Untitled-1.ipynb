```python
import requests
import json
import time
from confluent_kafka import Producer
```

```python
# Cấu hình Kafka
kafka_config = {
    'bootstrap.servers': 'localhost:9092',  # Địa chỉ Kafka broker
    'client.id': 'weather-producer'
}
producer = Producer(kafka_config)

# Thông tin API
api_url = "https://api.weatherapi.com/v1/history.json"
api_key = "be31cf794df34c0b848162013242211"
location = "Hanoi"
```

```python
# Hàm gửi yêu cầu đến API và xử lý dữ liệu
def fetch_filter_and_publish(date, hour):
    try:
        # Thiết lập tham số cho yêu cầu API
        params = {
            "key": api_key,
            "q": location,
            "dt": date  # Ngày cần truy vấn
        }
        # Gửi yêu cầu đến API
        response = requests.get(api_url, params=params)
        response.raise_for_status()  # Kiểm tra lỗi HTTP
        data = response.json()

        # Kiểm tra dữ liệu
        if (
            'forecast' in data and
            'forecastday' in data['forecast'] and
            len(data['forecast']['forecastday']) > 0 and
            'hour' in data['forecast']['forecastday'][0] and
            len(data['forecast']['forecastday'][0]['hour']) > hour
        ):
            # Truy xuất dữ liệu từng giờ
            hourly_data = data['forecast']['forecastday'][0]['hour'][hour]
            filtered_data = {
                'datetime': hourly_data['time'],
                'name': data['location']['name'],
                'country': data['location']['country'],
                'latitude': data['location']['lat'],
                'longitude': data['location']['lon'],
                'temp_c': hourly_data['temp_c'],  # Nhiệt độ
                'wind_mph': hourly_data['wind_mph'],  # Tốc độ gió
                'humidity': hourly_data['humidity'],  # Độ ẩm
                'precip_mm': hourly_data['precip_mm'],  # Lượng mưa
                'condition': hourly_data['condition']['text'],  # Tình trạng thời tiết
            }
            print(json.dumps(filtered_data, indent=4))
            # Gửi dữ liệu lên Kafka
            producer.produce('weatherkafkatopic', json.dumps(filtered_data))
            print(f"Published data for {date} {hour}:00 to Kafka topic: weatherkafkatopic")
        else:
            print(f"No data available for {date} {hour}:00")
    except Exception as e:
        print(f"Error fetching data for {date} {hour}: {e}")

```

```python
# Lặp qua từng ngày và từng giờ để lấy dữ liệu
start_date = time.strptime("2024-07-02", "%Y-%m-%d")
end_date = time.strptime("2024-08-02", "%Y-%m-%d")  # Thay đổi ngày kết thúc theo ý muốn
current_date = start_date

while current_date <= end_date:
    date_str = time.strftime("%Y-%m-%d", current_date)
    print(f"Fetching data for date: {date_str}")

    # Lặp qua từng giờ trong ngày
    for hour in range(24):
        fetch_filter_and_publish(date_str, hour)
        #time.sleep(1)  # Để tránh vượt giới hạn API

    # Chuyển sang ngày tiếp theo
    current_date = time.localtime(time.mktime(current_date) + 86400)  # Thêm 1 ngày (86400 giây)

producer.flush()  # Đảm bảo tất cả các dữ liệu đã được gửi
```
